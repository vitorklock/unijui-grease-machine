# Grease Machine — Simulation

This notebook is the entry point for the grease machine simulation.

## Project Structure

```
grease_machine/          # Real control software (independent package)
  interfaces.py          # IMotor, IScale, IThermometer
  calibration/           # CalibrationPoint, CalibrationStore, Interpolator
  controllers/
    manual.py            # ManualController (motor_on / motor_off)
    automatic/
      controller.py      # AutomaticController.dispense()
      calibration_procedure.py  # CalibrationProcedure.run()

simulation/              # Simulation (uses grease_machine as library)
  hardware/              # Simulated IMotor, IScale, IThermometer
  scenarios/
    calibration_scenario.py    # Calibrate at 3 temperatures → plots + saves JSON
    dispensing_scenario.py     # Accuracy at various target weights
    manual_vs_auto_scenario.py # Human skill vs AutomaticController
```
